```python
# Reading Data

gdp_data=pd.read_excel('./Q1/gdp_data.xls', skiprows=3)
mal_nutrition_data=pd.read_excel('./Q1/malnutrition_data.xls', sheet_name='Data', skiprows=3)
metadata=pd.read_excel('./Q1/./malnutrition_data.xls', sheet_name=1)
```

```python
gdp_data2=gdp_data.drop(['Indicator Name', 'Indicator Code','Country Name'], axis=1)
mal_nutrition_data2=mal_nutrition_data.drop(['Indicator Name', 'Indicator Code','Country Name'], axis=1)

mal_nutrition_data2.set_index('Country Code', inplace=True)
gdp_data2.set_index('Country Code', inplace=True)

plt.figure(figsize=(10, 8), dpi=80)
plt.scatter(gdp_data2,mal_nutrition_data2)
plt.title("Malnutrition prevelence against GDP per Capita", color= 'green')
plt.xlabel("GDP", color='blue')
plt.ylabel("Malnutrition", color='blue')
#plt.show()
```

```python
 # Question 2 redesigned and chose what is right from above one and this

# get data of regions from metadata
regions=metadata['Region'].unique()
regions=list(regions)
regions.pop(1)
regions.pop(-1)
#  extract on country code and regions
data_from_metadata=metadata[['Country Code','Region']]

# # copy gdp data in gdp_data_with_reegion to merge with region columns
gdp_data_with_region=gdp_data
gdp_data_with_region=gdp_data_with_region.reset_index()
gdp_data_with_region=pd.merge(gdp_data_with_region,data_from_metadata,left_on='Country Code', right_on='Country Code')
# # #  Drop unwanted cols
gdp_data_with_region=gdp_data_with_region.drop(columns=['Country Name','Indicator Name', 'Indicator Code','index'])
# # #  Set index again
gdp_data_with_region.set_index('Country Code', inplace=True)

# # #  do the same job for malnutrition data
mal_nutrition_data_with_regions=mal_nutrition_data
mal_nutrition_data_with_regions=mal_nutrition_data_with_regions.reset_index()
mal_nutrition_data_with_regions=pd.merge(mal_nutrition_data_with_regions, data_from_metadata, left_on='Country Code', right_on= 'Country Code')
mal_nutrition_data_with_regions=mal_nutrition_data_with_regions.drop(columns=['Country Name','Indicator Name', 'Indicator Code', 'index'])
mal_nutrition_data_with_regions.set_index('Country Code', inplace=True)

# # #  Now got data, I begin preparatio for plotting a graph
gdp_data_region=list()
malnutrition_data_region=list()
for i in regions:
    gdp_data_region.append(gdp_data_with_region[gdp_data_with_region['Region'] == i].drop(columns=['Region']))
    malnutrition_data_region.append(mal_nutrition_data_with_regions[mal_nutrition_data_with_regions['Region'] == i].drop(columns=['Region']))

# # # Time to plot my graph
colors=['red', 'blue', 'green', 'purple', 'orange', 'yellow']
plt.figure(figsize=(10, 8), dpi=80)

for data_gdp, data_mal, col, region in zip(gdp_data_region, malnutrition_data_region, colors, regions):
    plt.scatter(data_gdp,data_mal, color=col, label=region)
    
plt.xlabel("GDP", color='blue')
plt.ylabel("Malnutrition", color='blue')
plt.legend()
plt.title("malnutrition prevalence against GDP per capita shown by geographical regions  \n", color='green')
plt.legend()
plt.show()
```

```python
# get income level from metadata
income_levels=metadata['IncomeGroup'].unique().tolist()
income_levels.pop(1)
data_from_metadata=metadata[['Country Code','IncomeGroup']]

# process to merge income level with gdp data and store in a new variable
gdp_data_with_income_levels=gdp_data
# gdp_data_with_income_levels=gdp_data.drop(columns=['regions'])
gdp_data_with_income_levels=gdp_data_with_income_levels.reset_index()
gdp_data_with_income_levels=pd.merge(gdp_data_with_income_levels, data_from_metadata, left_on='Country Code', right_on='Country Code')
gdp_data_with_income_levels=gdp_data_with_income_levels.drop(columns=['Country Name', 'Indicator Name', 'Indicator Code', 'index'])
gdp_data_with_income_levels.set_index('Country Code', inplace=True)

# # Merge income levels to malnutriotion using the same process
# # # merge income level with malnutrition data and store in a new variable
mal_nutrition_data_with_income_levels=mal_nutrition_data
# mal_nutrition_data_with_income_levels=mal_nutrition_data_with_income_levels.drop(columns=['regions'])
mal_nutrition_data_with_income_levels=mal_nutrition_data_with_income_levels.reset_index()
mal_nutrition_data_with_income_levels=pd.merge(mal_nutrition_data_with_income_levels, data_from_metadata, left_on='Country Code', right_on='Country Code')
mal_nutrition_data_with_income_levels=mal_nutrition_data_with_income_levels.drop(columns=['Country Name','Indicator Name', 'Indicator Code','index'])
mal_nutrition_data_with_income_levels.set_index('Country Code', inplace=True)

# # setting indeces to new dataframes

colors=['red', 'green', 'blue', 'yellow']
plt.figure(figsize=(10, 8), dpi=80)

gdp_data_in_income_level=list()
malnutrition_data_in_income_level=list()
i=0
for level, colr in zip(income_levels,colors):
    gdp_data_in_income_level.append(gdp_data_with_income_levels[gdp_data_with_income_levels['IncomeGroup'] == level].drop(columns=['IncomeGroup']))
    malnutrition_data_in_income_level.append(mal_nutrition_data_with_income_levels[mal_nutrition_data_with_income_levels['IncomeGroup'] == level].drop(columns=['IncomeGroup']))
    plt.scatter(gdp_data_in_income_level[i], malnutrition_data_in_income_level[i], label=level, color=colr)
    i+=1
        
plt.xlabel("GDP", color='blue')
plt.ylabel("Malnutrition", color='blue')
plt.legend()
plt.title("malnutrition prevalence against GDP per capita marked by Income Levels \n", color='green')
plt.legend()
plt.show()
```

```python
import numpy as np

#  Get data from fertility 1990
fertility_data=pd.read_excel('./Q4/fertility_data.xls', skiprows=3)
fertility_1990_2010=fertility_data[['Country Name','1990','2010']]
fertility_2010=fertility_1990_2010['2010']
fertility_1990=fertility_1990_2010['1990']

#  I am gon to use a formula for cumulative distribution
N=len(fertility_1990)
X1= np.sort(fertility_1990)
F1 = np.array(range(N))/float(N)

N1=len(fertility_2010)
X2=np.sort(fertility_2010)
F2=np.array(range(N1))/float(N1)

#  find mean and median of data of 1990 and 2010
mean_1990=fertility_1990.mean()
median_1990=fertility_1990.median(axis=0)

mean_2010=fertility_2010.mean()
median_2010=fertility_2010.median(axis=0)

#  plotting the graph and put all asked data
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(X1, F1, label='Fertility rate variable in 1990')
plt.plot(X2, F2, label='Fertility rate variable in 2010')
plt.axvline(x = mean_1990,color = 'r', label = 'Mean of data in 1990')
plt.axvline(x = median_1990,color = 'b', label = 'Median of data in 1990')
plt.axvline(x = mean_2010,color = 'g', label = 'Mean of data in 2010')
plt.axvline(x = median_2010,color = 'purple', label = 'Median of data in 2010')
plt.legend()
plt.title('Cumulative distribution for the fertility rate variable in 1990 and 2010\n', color='b')
plt.show()

print("Mean and Median Data:")
print("Mean in 1990:  ",mean_1990)
print("Median in 1990:",median_1990)
print("Mean in 2010:  ",mean_2010)
print("Median in 2010:",median_2010)
```
