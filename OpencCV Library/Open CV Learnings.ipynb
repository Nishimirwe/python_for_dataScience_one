# Get Video from Camera using opencv

```python
import cv2
```

```python
camera=cv2.VideoCapture(0)
```

```python
while True:
    rt, frame= camera.read()
    if rt==True:
        cv2.imshow("My video capture", frame)
        key_pressed=cv2.waitKey(1) & 0xFF
        if key_pressed==ord('q') or key_pressed==ord('Q'): # This line will quit loop if q or Q is pressed. Carefully
            break
    else:
        continue
        
camera.release()
cv2.destroyAllWindows()
```

# FACE RECOGNITION WITH OPENCV

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=plt.imread('./images/group.jpg')
plt.axis('off')
plt.imshow(img)
```

```python
detector=cv2.CascadeClassifier('./haarcascade/haarcascade_frontalface_default.xml')
```

```python
all_faces = detector.detectMultiScale(img)
all_faces
```

```python
x,y,w,h=all_faces[0]
```

```python
x,y,w,h
```

```python
picture=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
```

```python
plt.imshow(picture)
```

```python
img=plt.imread('./images/group.jpg')
```

```python
def detectFaces(img):
    x=0
    y=0
    w=0
    h=0
    detector=cv2.CascadeClassifier('./haarcascade/haarcascade_frontalface_default.xml')
    all_faces=detector.detectMultiScale(img)
    pic=img
    for i in all_faces:
        x,y,w,h=i
        cv2.rectangle(pic,(x,y),(x+w,y+h),(0,255,0),2)
    return pic
```

```python
detectedImg=detectFaces(img)
```

```python
plt.imshow(img)
plt.show()
```
