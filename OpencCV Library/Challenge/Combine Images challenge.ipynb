```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img1=plt.imread('./images/img1.jpg')
img2=plt.imread('./images/img2.jpg')
img3=plt.imread('./images/img3.jpg')
img4=plt.imread('./images/img4.jpg')
img5=plt.imread('./images/img5.jpeg')
```

```python
plt.imshow(img1)
```

```python
img1=cv2.resize(img1,(200,200))
img2=cv2.resize(img2,(200,200))
img3=cv2.resize(img3,(200,200))
img4=cv2.resize(img4,(200,200))
img5=cv2.resize(img5,(100,100))
```

```python
# create another bigger Image to contain all images
full_image=np.zeros((430,430,3), dtype='uint8')
```

```python
plt.imshow(img1)
```
