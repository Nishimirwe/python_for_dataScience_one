- it's a parser for html and xml

```python
# !pip install bs4
```

```python
import bs4
import requests
import pandas as pd
import re
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_and_their_capitals_in_native_languages"
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
text = res.content
```

```python
soup = bs4.BeautifulSoup(text)
```

```python
type(soup)
```

```python
soup.title
```

```python
soup.h1.text
```

```python
soup.find('h1')
```

```python
all_tables = soup.find_all('table', attrs = {"class" : "wikitable"})
```

```python
len(all_tables)
```

### getting information about one country

```python
D_table = all_tables[3].find_all('tr')[1:]
```

```python
len(D_table)
```

```python
all_tds = D_table[0].find_all('td')
```

```python
len(all_tds)
```

```python
all_tds[0].img.attrs['src']
```

```python
country = all_tds[0].text
capital = all_tds[1].text
language = all_tds[-1].text
```

```python
print(country, capital, language)
```

```python
df = pd.DataFrame({}, columns=['Country', 'Capital', 'Language'])
```

```python
df
```

### getting information from all the tables

```python
for one_table in all_tables:
    all_rows= one_table.find_all('tr')[1:]
    for tr in all_rows:
        all_tds = tr.find_all('td')

        country = all_tds[0].text
        capital = all_tds[1].text
        language = all_tds[-1].text

        ele = {'Country': country, 'Capital' : capital, 'Language' : language}
        
        df = df.append(ele, ignore_index=True)
        
        print(country,"==",capital,"==",language)
        print()
```

```python
df.shape
```

```python
df.head()
```

```python
df['Language']  = df['Language'].apply(lambda x : x[:-1])
```

```python
df.to_csv("countries.csv", index=False)
```

```python
df['Country'] =  df['Country'].apply(lambda x : re.sub('\[.*\]', "", x.strip()))
```
