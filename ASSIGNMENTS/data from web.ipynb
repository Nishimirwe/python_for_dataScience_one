```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel('./files/ebola_download.xls')
```

```python
data100=data[(data['Cases']>=100) & (data['Death']>=100)]
data500=data[(data['Cases']>=500) & (data['Death']>=500)]
data1000=data[(data['Cases']>=1000) & (data['Death']>=1000)]
data2000=data[(data['Cases']>=2000) & (data['Death']>=2000)]
data5000=data[(data['Cases']>=5000) & (data['Death']>=5000)]
```

```python
startDate5000=list(data5000['Date'])
stDate5000=str(startDate5000[0])[:-9]

endDate5000=list(data5000['Date'])
stEndDate5000=str(startDate5000[-1])[:-9]


startDate100=list(data100['Date'])
stDate100=str(startDate100[0])[:-9]

startDate500=list(data500['Date'])
stDate500=str(startDate500[0])[:-9]

startDate1000=list(data1000['Date'])
stDate1000=str(startDate1000[0])[:-9]

startDate2000=list(data2000['Date'])
stDate2000=str(startDate2000[0])[:-9]

dates=[stDate100,stDate500,stDate1000,stDate2000,stDate5000,stEndDate5000]

startCase100=list(data100['Cases'])[0]
startCase500=list(data500['Cases'])[0]
startCase1000=list(data1000['Cases'])[0]
startCase2000=list(data2000['Cases'])[0]
startCase5000=list(data5000['Cases'])[0]
endCase5000=list(data5000['Cases'])[-1]
cases=[startCase100,startCase500,startCase1000,startCase2000,startCase5000,endCase5000]

startDeath100=list(data100['Death'])[0]
startDeath500=list(data500['Death'])[0]
startDeath1000=list(data1000['Death'])[0]
startDeath2000=list(data2000['Death'])[0]
startDeath5000=list(data5000['Death'])[0]
endDeath5000=list(data5000['Death'])[-1]
deaths=[startDeath100,startDeath500,startDeath1000,startDeath2000,startDeath5000,endDeath5000]
```

```python
allDates=data['Date']
allCases=data['Cases']

plt.plot(allDates,allCases)
plt.show()
```

```python
plt.plot(dates,cases, color="red", label="Cases", marker='o')
plt.plot(dates,deaths, color="blue", label="Deaths", marker='o')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
