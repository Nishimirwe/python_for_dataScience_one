# Bar Graph

```python
x=np.arange(5)
```

```python
car1=[89,42,70,24,35]
car2=[56,15,3.5,28,33]
```

```python
plt.bar(x,car2, color='purple', label='Toyota', width=0.4, tick_label=[2016,2017,2018,2019,2020])
plt.bar(x+0.4,car1,color='blue', label='Benz',width=0.4,)
plt.legend()
plt.title("Comparison")

plt.show()
```
